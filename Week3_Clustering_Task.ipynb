## Data Preprocessing
Do what you think you need such as:
* Remove the outliers
* Impute missing data
* Scale the data
* Reduce dimentions using PCA
* Implement One-Hot Encoding for nominal categorical variables.

```python

#Remove the outliers
#هنا انا راح احذف البيانات الي راح تاثر علي بالقيم النهائيه وحذها ماراح ياثر علي
#  outliersراح استخدم قانون

def remove_outliers(df,column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound  = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    return df.loc[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# هنا راح اسوي فور لوب وتحذف القيم ب استثناء الاوبجكت
for i in df.select_dtypes(exclude=['object']).columns:
    df = remove_outliers(df,i)
```

```python
df.isnull().sum()
```

```python
# missing data,delet colum
# بعد ماسويت اوت لاير راح القيم الفارغه ف ماراح احتاج اسوي ديليتد

#df.drop(columns =['Accident_Severity'])
#df.drop(columns =['Urban_or_Rural_Area'])
#df.drop(columns=['Light_Conditions'])

# هنا سويت فور لوب بحيث يدور على كل الاعمده ويعبي القيم الفارغه ب المود

null_columns = df
for i in null_columns:
    df[i] = df[i].fillna(df[i].mode()[0], inplace=True)
```

```python
##Encode CV convert object into numerc

le = LabelEncoder()
df['Date'] = le.fit_transform(df['Date'])
df['Local_Authority_(Highway)'] = le.fit_transform(df['Local_Authority_(Highway)'])
df['Road_Type'] = le.fit_transform(df['Road_Type'])
df['Pedestrian_Crossing-Human_Control '] = le.fit_transform(df['Pedestrian_Crossing-Human_Control'])



ss = StandardScaler()
df = pd.DataFrame(ss.fit_transform(df),columns=df.columns)
```
